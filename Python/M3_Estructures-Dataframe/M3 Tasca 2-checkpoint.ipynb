#### Altres calculs rellevants
1. Agrupar retrassos per mesos
2. Agrupar per dies de la setmana (Canviar el número pel dia)
3. Mirar quina es la major causa dels retrassos


```python
# Order by the Months of the mean of ArrDelay is higher 
mesos = vols_reduit_net.groupby('Month')['ArrDelay'].mean()
mesos = vols_reduit_net.groupby('Month').agg({'ArrDelay': 'mean'})

mesos.sort_values('ArrDelay', ascending=False)
```

Veiem que els mesos amb mes retrassos son desembre i juny i juliol (sorpren que no hi surti agost). Em sembla interessant saber quins mesos hi ha més retrassos per si s'ha de fer algún tipus de reforç.

```python
# Order by the DayOfWeek of the mean of ArrDelay is higher
# Change values of the column to name of the day.
vols_reduit_net['DayOfWeek'].replace({1: "Monday", 2: "Tuesday", 3: "Wednesday", 4: "Thursday", 5: "Friday", 6: "Saturday", 7: "Sunday"}, inplace=True)

dies_setmana = vols_reduit_net.groupby('DayOfWeek')['ArrDelay'].mean()
dies_setmana = vols_reduit_net.groupby('DayOfWeek').agg({'ArrDelay': 'mean'})

dies_setmana.sort_values('ArrDelay', ascending=False)
```

No sembla que hi hagi una gran diferència entre els diferents dies de la setmana.

```python
# Check which is the major cause of the delay
Qty_DepDelay = (vols_reduit_net['DepDelay'] > 0).sum()
Qty_ArrDelay = (vols_reduit_net['ArrDelay'] > 0).sum()
Qty_CarrierDelay = (vols_reduit_net['CarrierDelay'] > 0).sum()
Qty_WeatherDelay = (vols_reduit_net['WeatherDelay'] > 0).sum()
Qty_NASDelay = (vols_reduit_net['NASDelay'] > 0).sum()
Qty_SecurityDelay = (vols_reduit_net['SecurityDelay'] > 0).sum()
Qty_LateAircraftDelay = (vols_reduit_net['LateAircraftDelay'] > 0).sum()

print("El total de vols amb retràs a l'enlairament son:", Qty_DepDelay)
print("El total de vols amb retràs en l'aterratge son:", Qty_ArrDelay)
print("El total de vols amb retràs degut el temps son:", Qty_WeatherDelay)
print("El total de vols amb retràs degut a la gestió del trànsit son:", Qty_NASDelay)
print("El total de vols amb retràs degut a la seguretat son:", Qty_SecurityDelay)
print("El total de vols amb retràs degut al vol anterior son:", Qty_LateAircraftDelay)
```

Podem comprovar que el LateAircraft i el NAS son els causants de la majoria de retrassos. Per contrapartida la seguretat i el temps (mes dificil de controlar) son els causants menys habituals.
